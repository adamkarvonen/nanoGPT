```python
test_equations = []

with open("data/addition/test.txt", "r") as f:
    for line in f:
        split_str = line.split('+')
        first_num = int(split_str[0])
        second_num = int(split_str[1].split('=')[0])
        test_equations.append([f"{first_num}+{second_num}=", (first_num + second_num)])
print(test_equations)
```
